```python
import torch

a = list(range(9))
print('List a = ', a)
```

```python
tensor_a = torch.tensor(a)
print('tensor a = ', tensor_a)
# How to check Tensor Size
print("Size of tensor: ",tensor_a.size())
print("Shape of tensor: ", tensor_a.shape)
print("Offset of tensor: ",tensor_a.storage_offset())
print("Stride of tensor: ",tensor_a.stride())

tensor_b = torch.rand(2, 5, 3)
print(tensor_b)
print("Size of tensor b: ",tensor_b.size())
print("Shape of tensor b: ", tensor_b.shape)
print("Offset of tensor b: ",tensor_b.storage_offset())
print("Stride of tensor: ",tensor_b.stride())
```

```python
b = tensor_a.view(3,3)
print('tensor b: ',b )
print("Size of tensor b: ", b.size())
print("Shape of tensor b: ", b.shape)
print("Offset of tensor b: ", b.storage_offset())
print("Stride of tensor: ", b.stride())
```

```python
c = b[1:, 1:]
print('size of c: ', c.shape)
print('offset of c: ', c.storage_offset())
print('stride of c: ', c.stride())
```

```python
import math

angle_rad = torch.tensor(math.pi / 3)  # 60 degree Radian
cosine_value = torch.cos(angle_rad)

print('radian 60 = ', angle_rad.item())
print('cos 60 =', cosine_value)
```

```python
tensor_angles = torch.tensor([0.0, math.pi / 2, math.pi, 3* math.pi / 2, 2 * math.pi])
print('angles = ', tensor_angles)
cosines = torch.cos(tensor_angles)
print('Cosine of angles = ', cosines)
```
